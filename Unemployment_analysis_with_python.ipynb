```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## 2. Load Data

```python
df=pd.read_csv("Unemployment_Rate_upto_11_20201.csv")
```

```python
df.head()
```

At this step we have to check is there any column has empty dataset.

```python
df.isnull().sum()
```

```python
df.info()
```

By df.info() we see that then columns name are not correct. so first we need to correct column's name.

```python
df.columns=["States","Date","Frequency","Estimated Unemployment Rate (%)","Estimated Employed","Estimated Labour Participation Rate (%)","Region","longitude","latitude","Addition of Estimated Unemployment Rate (%)"]
df
```

## 3. Analysis the data


```python
df.corr()
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr())
plt.show()
```

```python
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate (%)",hue="Region", data=df)
plt.show()
```

```python
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=df)
plt.show()
```

```python
fig = plt.figure(figsize = (30, 10))

# creating the bar plot
plt.xticks(rotation = 90)
plt.xticks(fontsize=25)
plt.yticks(fontsize=20)
plt.bar(df["States"], df["Addition of Estimated Unemployment Rate (%)"], color =["#2e2b28","#d7e1ee", "#cbd6e4", "#bfcbdb", "#b3bfd1", "#a4a2a8", "#df8879", "#c86558", "#b04238", "#991f17"],width = 0.4)
plt.title("Total  Estimated Unemployment Rate (%) from 31-01-2020 to 31-10-2020",fontsize=50)
plt.xlabel("States",fontsize=25)
plt.ylabel("Unemployment Rate in (%)",fontsize=25)
#
```

```python
fig = plt.figure(figsize = (30, 10))

# creating the bar plot
plt.xticks(rotation = 90)
plt.xticks(fontsize=25)
plt.yticks(fontsize=20)
plt.bar(df["States"], df["Estimated Unemployment Rate (%)"], color ="#333333",width = 0.4)
plt.title("Estimated Unemployment Rate (%) from 31-01-2020 to 31-10-2020",fontsize=50)
plt.xlabel("States",fontsize=25)
plt.ylabel("Unemployment Rate in (%)",fontsize=25)
#
```

```python
unemploment = df[["States", "Region", "Estimated Unemployment Rate (%)"]]
figure = px.sunburst(unemploment, path=["Region", "States"], 
                     values="Estimated Unemployment Rate (%)", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```
